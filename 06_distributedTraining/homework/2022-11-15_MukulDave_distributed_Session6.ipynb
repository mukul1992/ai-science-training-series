The total training time increases with number of GPUs rather than decreasing, indicating that the communication overhead is more dominant than the speedup from division of work.

The accuracy drops drastically for np=8. While scaling learning rate with number of processes makes sense because we have more information in each batch, there is still risk of taking too large a step in each batch and falling into a local minimum or a low gradient region in the feature space, especially at the start of the training.

```python
np = [1, 2, 4, 8]
total_times = [29.5, 41.7, 42.5, 53.2]

plt.plot(np,total_times)
plt.xlabel("Number of GPUs")
plt.ylabel("total training time (s)")
plt.show()
```
